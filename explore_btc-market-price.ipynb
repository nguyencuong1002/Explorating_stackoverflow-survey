```python
df = pd.read_csv('./data/btc-market-price.csv', header=None)
df
```

```python
df.columns = ['Timestamp', 'Revenue']
df
```

```python
# formart date time
df['Timestamp'] = pd.to_datetime(df['Timestamp']) 
arr_day = []
arr_month = []
arr_year = []
for i in range(df.count()['Timestamp']):
    arr_day.append(str(df['Timestamp'][i])[8:10])
    arr_month.append(str(df['Timestamp'][i])[5:7])
    arr_year.append(str(df['Timestamp'][i])[0:4])
df['day'] = arr_day    
df['month'] = arr_month
df['year'] = arr_year
df
```
